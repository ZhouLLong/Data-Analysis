# Practise 3

Suppose you want to normalize each student's grade based on their age.
1. Write a function using Python Pandas that will add a new column to your containing a new grade normailzed against the mean age of students.

```python
d = {'Age': [20, 19,22,21], 
     'Favorite Color': ['blue', 'blue','red','yellow'],
    'Grade':[88,95,92,70],
    'Name':['Tom','Jerry','Cat','Rat']}
df_nor = pd.DataFrame(data=d)
df_nor
```

```python
c_mean = df_nor['Age'].mean()
c_mean
```

```python
c_std = df_nor['Age'].std()
c_std
```

```python
df_nor['z_score'] = (df_nor['Age'] - c_mean)/c_std
df_nor
```

```python
df_nor['normailzed'] = df_nor['Grade'] + (df_nor['Grade'].std() * df_nor['z_score'])
df_nor
```

# Practise 4

Suppose a company is taxed 20% of earnings for a given quarter are 3M USD.
If lower than 3M USD, the company  is taxed at a lower rate of 15%.

Write a function using Python to calculate both the pre-tax and post-tax earnings for a given company, with the ablity to feed in the # of widgets sold as well as the average price of the widget as inputs(# of widgets sold*avg. price of widgets = total earnings in this case).

```python
def earnings_cal(widgets,cost):
    pre_tax = widgets*cost
    
    if pre_tax >= 3000000:
        post_tax = pre_tax* 0.8
    else:
        post_tax = pre_tax* 0.85
    
    return pre_tax,post_tax
```

## Practise 6

You're given a set of data that is aggregated on a monthly basis (all illustrated in Table A).Can you write code that can expand this monthly table into a daily table which spreads revenue across 30 day period(as shown in Table B)?

```python
df_reven = pd.DataFrame(data= {'Month': [1, 2,3,],
            
     'Revenue': [300, 330,390]

    })
df_reven.info()
df_reven
```

```python
def table_expander(df_reven,month_length):
    new_dataframe = []
    for i in range(month_length):
        new_dataframe.append(df_reven)
    
    output = pd.concat(new_dataframe)
    
    output['Revenue'] = output['Revenue']/ month_length
    
    output = output.sort_values('Month')
    
    return output
```

```python
table_expander(df_reven,31)
```

```python
temp= table_expander(df_reven,31)
temp[temp['Month']==3].head()
```
