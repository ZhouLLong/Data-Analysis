### 're' function making it more easier and super powerful

```python
import re
df.loc[~df['Name'].str.contains('Mega')].head(15)
```

## Aggregate Statistics & Groupby

```python
df = pd.read_csv('modified.csv')
```

```python
df.groupby(['Type 1']).mean().sort_values('Attack', ascending = False)
df.groupby(['Type 1']).mean().sort_values(['HP','Attack','Defense'], ascending=[0,0,0])
```

```python
df.groupby(['Type 1']).agg({'Defense':'mean','Speed':'mean'})
```

```python
df = pd.read_csv('modified.csv')

df['count'] = 1

df.groupby(['Type 1', 'Type 2']).count()['count']
```

```python
new_df = pd.DataFrame(columns=df.columns)

for df in pd.read_csv('modified.csv', chunksize=5):
    results = df.groupby(['Type 1']).count()
    
    new_df = pd.concat([new_df, results])
```
