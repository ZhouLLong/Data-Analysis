```python
import pandas as pd
import warnings
```

```python
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("data_cleaning_challenge.csv")
df.head(20)
```

```python
nan_df = df[df.isna().any(axis=1)]
nan_df.describe
```

#### Droping columns

```python
data_drop = df.drop(columns=["Unnamed: 9", "Unnamed: 10"])
data_drop.head(20)
```

```python
no_nan = data_drop[data_drop["Row Type"].notna()]
no_nan
```

### Iteration & count

```python
column_values = []
iteration = 0

for i in no_nan["Row Type"]:
    if "first name" in i:
        iteration += 1
    column_values.append(iteration)
```

```python
column_values
```

```python
iteration
```

```python
iter_column_added = no_nan
```

```python
iter_column_added["Iteration"] = column_values
iter_column_added.head(30)
```

```python
drop_extra_column_rows = iter_column_added[iter_column_added["Row Type"] != "Row Type"]
drop_extra_column_rows.head(30)
```

#### contains; rename; drop

```python
name_dataframe = drop_extra_column_rows.loc[drop_extra_column_rows[
    "Row Type"].str.contains("first name")]
name_dataframe.head(10)  # 5994rows total
```

```python
name_dataframe = name_dataframe.loc[:,["Row Type", "Iter Number", "Power1", "Iteration"]]
name_dataframe
```

```python
name_dataframe.rename(columns={"Row Type": "First Name",
                               "Iter Number":"Last Name", 
                               "Power1": "Date"}, inplace=True)
name_dataframe["First Name"] = name_dataframe["First Name"].str[12:]
name_dataframe["Last Name"] = name_dataframe["Last Name"].str[11:]
name_dataframe["Date"] = name_dataframe["Date"].str[6:]
```

```python
name_dataframe.head(15)
```

```python
no_name_dataframe = drop_extra_column_rows[
    ~drop_extra_column_rows["Row Type"].str.contains("first name")]
no_name_dataframe.head(15)
```

```python
no_name_dataframe
```

## Merging (pd.merge)

```python
df_mer = pd.merge(left=name_dataframe, right=no_name_dataframe, how="inner", on="Iteration")
df_clean = df_mer.drop(columns=["Iteration"])
```
